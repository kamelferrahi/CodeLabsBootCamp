```python
import pandas as pd
from sklearn.model_selection import train_test_split
from scipy import stats
import numpy as np
```

```python
df = pd.read_csv("food-allergy-analysis-Zenodo.csv")
```

```python
df
```

```python

#newdf contains only the numerical column of df
#convert categorical data into numerical one
newdf=pd.get_dummies(df, columns=["GENDER_FACTOR", "RACE_FACTOR","ETHNICITY_FACTOR","PAYER_FACTOR","ATOPIC_MARCH_COHORT"], prefix=["GENDER", "RACE","ETHNICITY","PAYER","ATOPIC"])

newdf
```

```python
#there a lot of Nan so lets fill them
newdf = newdf.fillna(newdf.mean())
newdf.isnull().sum()
newdf
```

```python
#There are still Nan Value in TreeNut-ALG-END 
newdf.pop("TREENUT_ALG_END")
#Lets remove the feature that works with the previous one
newdf.pop("TREENUT_ALG_START")
#Remove Useless feature
newdf.pop("BIRTH_YEAR")


newdf.isnull().sum()
```

```python
#Lets start prediciting food allerggies!
x=newdf.iloc[:,:-1].values
y=newdf.iloc[:,-1:].values
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
sk_model=LogisticRegression()
```

```python
# train the model
sk_model.fit(x_train, y_train)
```

```python
#get accuracy of the model
def get_accuracy(y_predict,y_test):
    correctly_classified=0
    for count in range(np.size(y_predict)):
        if y_test[count]==y_predict[count]:
            correctly_classified=correctly_classified+1
    acc=(correctly_classified/len(y_predict))*100
    print("accuracy on testset={:.2f}".format(acc))
    return acc
 
```

```python
y_predict=sk_model.predict(x_test)
```

```python
accur=get_accuracy(y_predict,y_test)
```
