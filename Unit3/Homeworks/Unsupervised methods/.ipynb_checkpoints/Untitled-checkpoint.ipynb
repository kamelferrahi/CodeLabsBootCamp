```python
import pandas as pd
```

```python
df = pd.read_csv("clustering_data.csv")
print("Columns: ",df.keys())
print("Size: ",df.size)
print("Info: ",df.info())
```

```python
import matplotlib.pyplot as plt
from  sklearn.cluster  import KMeans
```

```python
x = df[df.keys()[0]]
y = df[df.keys()[1]]
plt.scatter(x,y)
```

```python
#lets use the elbow method
values =[]
for k in range(2,20):
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(df.values)
    values.append(kmeans.inertia_)
```

```python
fig = plt.figure(figsize=(15,5))
plt.plot(range(2,20),values)
plt.grid(True)
```

```python
#so by the elbow method the numbr of cluster is 6
model = KMeans(n_clusters = 6, init ="random" ,random_state = 1)
y_predic = model.fit_predict(df.values)
```

```python
print(y_predic)
```

```python
plt.scatter(x,y,c=y_predic)
```
