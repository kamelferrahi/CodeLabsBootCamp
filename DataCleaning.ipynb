```python
import pandas as pd
```

```python
df = pd.read_csv("videosUS.csv")
print("Number of missing value in each column:")

print(pd.isnull(df).sum())
```

```python
#Get the sum of all column then sum it all you have the number of missing value in all the dataframe 
#only add the column that have more then 0 misisng value in the string results
columns = pd.isnull(df).sum()
result = ""
num_missing = 0
for key in columns.keys():
    num_missing += columns[key]
    if (columns[key] != 0): result +=" "+key

result = result[1:]
print("The number of missing value: ",num_missing)
print("\nThe columns that contains missing value are: "+result)
```

```python
numerics = ['int16', 'int32', 'int64', 'float16', 'float32', 'float64']
#newdf contains only the numerical column of df
newdf = df.select_dtypes(include=numerics)
print(newdf)
```

```python
#for each column calculate the mean median and quartiles
print("Some value can have the same number of occurence in a column so that's why the mode isn't unique for a column")
for key in newdf.keys():
    print("The column "+key+" :")
    col = newdf[key]
    print("The median of the column: "+str(col.median()))
    print("The mean of the column: "+str(col.mean()))
    print("The first quantile of the column: "+str(col.quantile(q=0.25)))
    print("The second quantile of the column: "+str(col.quantile(q=0.5)))
    print("The third quantile of the column: "+str(col.quantile(q=0.75))+"\n")
    #mode = col.mode()
    #res = ""
    #for modes in mode:
        #res += " "+str(modes)
```

```python
#Calculates the outliers for each element in the newdf dataframe  with the z-score
for key in newdf.keys():
    print("\nOutliers of the column "+key+" :")
    d = newdf[key]
    std = d.std()
    m = d.mean()
    cpt = 0
    for x in d:
        z = (x-m)/std
        if (z>3):
            print(x)
            cpt+=1
    print("Number of outliers in the column:",cpt)
```

```python
keys = newdf.keys()[1:]
#calculate the zscore for each column then delete all the lines that containts outliers 
for key in keys:
    newdf["zscore"] = (newdf[key]- newdf[key].mean())/ newdf[key].std()
    newdf = newdf[newdf["zscore"]<=3]
    df[key] = newdf[key]
df.dropna(inplace=True)
```

```python
def most_common(di): #Function to get the key with the highest value in a dictionnary
    return max(di, key= lambda x: di[x])

```

```python
#get akk the tag in dictionnary with the key is the tags and the value is the number of occurence of that tag in the dataframe
tags = df["tags"]
di = dict()
for row in tags:
    for tag in row.split("|"):
        di[tag] = di.get(tag,0)+1  
        
print("There is",len(di.keys()),"tags")        
print("The most common tag is:",most_common(di))
```
